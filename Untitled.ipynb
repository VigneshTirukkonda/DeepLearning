```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('train.csv')
val = pd.read_csv('test.csv')

dan = data.copy(deep= True)
```

```python
def fill_missing():
    dset = [dan, val]
    
    for data_set in dset:
        data_set.Age.fillna(data_set.Age.median(), inplace= True)
        data_set.Embarked.fillna('S', inplace= True)
        data_set.Fare.fillna(data_set.Fare.median(), inplace= True)
        
fill_missing()        
```

```python
def add_features():
    dset = [dan, val]
    for data_set in dset:
        # Title
        data_set['Title'] = data_set.Name.apply(lambda x: x.split(', ')[1].split('.')[0])
        title_count = data_set['Title'].value_counts()
        title = title_count < 10
        data_set.Title = data_set.Title.apply(lambda x: 'Misc' if title.loc[x] == True else x)
        
        #Family size
        
add_features()
```

```python
def drop_unwanted(test):
    dset = [dan, val]
    
    if test == True:
        ret = []
        for data_set in dset:
            ret.append(data_set.drop(['PassengerId','Ticket','Cabin','Name'], axis= 1))
        return ret
    else:
        for data_set in dset:
            data_set.drop(['PassengerId','Ticket','Cabin','Name'], axis= 1, inplace= True)
        
drop_unwanted(test = False)
```
